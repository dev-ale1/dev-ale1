```
Uma variável guarda UM aluno. Mas o professor tem 30 alunos. Como guardar isso?
```

````python
# Lista de chamada da Kedu
diario_3_ano = ["Alex","Carlos","David","Enzo","Fabio","Gerson"]
# Para pegar um aluno específico, usamos a POSIÇÃO dele na fila.
# ATENÇÃO: O Python conta começando do ZERO (0).
# Isso é IMPORTANTÍSSIMO e sempre derruba iniciante.

primeiro_aluno_diario = diario_3_ano[0] # Pega Alex
Segundo_aluno_diario = diario_3_ano[1] # Pega Carlos
terceiro_aluno_diario = diario_3_ano[2] # Pega David
quarto_aluno_diario = diario_3_ano[3] # Pega Enzo
quinto_aluno_diario = diario_3_ano[4] # Pega Fabio
sexto_aluno_diario = diario_3_ano[5] # Pega Gerson

print("Chamando o numero 1:", primeiro_aluno_diario)
print("Chamando o numero 2:", Segundo_aluno_diario)
print("Chamando o numero 3:", terceiro_aluno_diario)
print("Chamando o numero 4:", quarto_aluno_diario)
print("Chamando o numero 5:", quinto_aluno_diario)
print("Chamando o numero 6:", sexto_aluno_diario)
````

```
🧱 FUNDAMENTO 3: O CÉREBRO DA KEDU (Condicionais - IF / ELSE)
Até agora o Python só obedece cegamente. Mas e se ele precisar PENSAR?
Exemplo: "Se o aluno tirou nota maior que 7, ele está APROVADO. Senão, está REPROVADO."

Para isso usamos o if (se) e else (senão).

Código 1.3:

**PONTO CRÍTICO (LEIA COM ATENÇÃO):**
Os espaços no começo da linha (identação) são OBRIGATÓRIOS em Python. Se você
esquecer o TAB, o código quebra. Isso é o que define o que está "dentro" do if.

```



````python
nota_do_pedro = 5.5

print("Analisando a nota do Pedro...")

if nota_do_pedro >= 7.0:
    # Tudo que está DENTRO do if precisa estar espaçado (TAB)
    print("✅ Status: APROVADO")
    print("Parabéns! Pode ir para as férias.")
else:
    # Tudo que está DENTRO do else também precisa de TAB
    print("❌ Status: REPROVADO")
    print("Precisa fazer recuperação.")

print("Fim da análise do sistema Unidade.")
````
